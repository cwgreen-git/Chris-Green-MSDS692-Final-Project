```python
terror = open('Terrorism2.txt')
terr_text = terror.read()
terror.close()
```

## Prepare Text

### Remove /n Line Breaks

```python
terr_text2 = terr_text.replace("\n", " ")#https://www.kite.com/python/answers/how-to-remove-all-line-breaks-from-a-string-in-python
```

### Text to Lower Case

```python
terr_text3 = terr_text2.lower()
```

### Remove Punctuation

```python
import re
terr_text4 = re.sub(r'[^\w\s]', '', terr_text3) #https://www.geeksforgeeks.org/python-remove-punctuation-from-string/
```

### Remove Digits

```python
terr_text5 = re.sub(r'[0-9]', '', terr_text4) #https://thispointer.com/python-remove-all-numbers-from-string/
```

### Remove Extra Spaces

```python
terr_text6 = re.sub(' +', ' ', terr_text5) #https://www.geeksforgeeks.org/python-remove-unwanted-spaces-from-string/
```

### Remove Stop Words

```python
from gensim.parsing.preprocessing import remove_stopwords #https://stackabuse.com/removing-stop-words-from-strings-in-python/
terr_text_final = remove_stopwords(terr_text6)
terr_text_final
```

## Word Cloud

```python
def plot_cloud(wordcloud):
    plt.figure(figsize=(40, 30))
    plt.imshow(wordcloud) 
    plt.axis("off");
wordcloud = WordCloud(width = 1000, height = 600, random_state=1, background_color='white', colormap='RdGy', collocations=False).generate(terr_text_final)
plot_cloud(wordcloud)
#https://towardsdatascience.com/simple-wordcloud-in-python-2ae54a9f58e5
```
